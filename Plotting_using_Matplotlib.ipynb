# Types of Data

- Numerical Data
- Categorical Data

- single column graph is called: `Univariate analysis`
- Two column graph is called: `bivariant analysis`
- More than two graph is called: `Multivariate analysis`

```python
# import the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('default')
```

### 2D Line plot

image.png

- Bivariate Analysis
- categorical -> numerical and numerical -> numerical
- Use case - Time series data

```python
# plotting a simple function
price = [48000,54000,57000,49000,47000,45000]
year = [2015,2016,2017,2018,2019,2020]

plt.plot(year,price)
```

```python
# from a pandas dataframe
batsman = pd.read_csv('/content/sharma-kohli.csv')
# batsman.head()
plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
# plotting multiple plots
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
# labels title
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.title('Rohit Sharma & Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Score')
```

```python
# colors(hex) and line(width and style) and marker(size)
plt.plot(batsman['index'],batsman['V Kohli'],color='green')  # can also use hex code
plt.plot(batsman['index'],batsman['RG Sharma'],color='black')
plt.title('Rohit Sharma & Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Score')
```

```python
# linestyle
plt.plot(batsman['index'],batsman['V Kohli'],color='green', linestyle='dashdot')  # solid, dashed, dotted, dasdot,
plt.plot(batsman['index'],batsman['RG Sharma'],color='black', linestyle='dashdot')
plt.title('Rohit Sharma & Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Score')
```

```python
# linewidth
plt.plot(batsman['index'],batsman['V Kohli'],color='green', linestyle='solid', linewidth=3)  
plt.plot(batsman['index'],batsman['RG Sharma'],color='black', linestyle='solid', linewidth=2)
plt.title('Rohit Sharma & Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Score')
```

```python
# marker and size
plt.plot(batsman['index'],batsman['V Kohli'],color='green', linestyle='solid', linewidth=3, marker='D', markersize=10)  # D, +, ., >, >, o
plt.plot(batsman['index'],batsman['RG Sharma'],color='black', linestyle='dashdot', linewidth=2, marker='o')
plt.title('Rohit Sharma & Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Score')
```

```python
# legend -> location
plt.plot(batsman['index'],batsman['V Kohli'],color='green', linestyle='solid', linewidth=3, marker='D', markersize=10, label='Virat') 
plt.plot(batsman['index'],batsman['RG Sharma'],color='black', linestyle='dashdot', linewidth=2, marker='o', label='Rohit')
plt.title('Rohit Sharma & Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Score')
plt.legend(loc='best') # default : best
	# upper right
	# upper left
	# lower left
	# lower right
	# right
	# center left
	# center right
	# lower center
	# upper center
	# center
```

```python
# limiting axes
price = [48000,54000,57000,49000,47000,45000,4500000]
year = [2015,2016,2017,2018,2019,2020,2021]

plt.plot(year,price)
# triming in y axis because of outlier
plt.ylim(0,75000)
# can also trim x axis
```

```python
# grid
plt.plot(batsman['index'],batsman['V Kohli'],color='green', linestyle='solid', linewidth=3, marker='D', markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'],color='black', linestyle='dashdot', linewidth=2, marker='o')
plt.title('Rohit Sharma & Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Run Score')
plt.grid()
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='green', linestyle='solid', linewidth=3, marker='D', markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'],color='black', linestyle='dashdot', linewidth=2, marker='o')
plt.title('Rohit Sharma & Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Run Score')
plt.grid()
plt.show()   # plt.show() is the main function to show us the plot
```

# Bar Chart
- Bivariate Analysis
- Numerical vs Categorical
- Use case - Aggregate analysis of groups

image.png

```python
# simple bar chart
children = [10,20,40,10,30]
colors = ['red','blue','green','yellow','pink']
plt.bar(colors, children, color='black')
```

```python
# horizontal bar chart
children = [10,20,40,10,30]
colors = ['red','blue','green','yellow','pink']
plt.barh(colors, children, color='black')
```

```python
# color and label
df = pd.read_csv('/content/batsman_season_record.csv')
df
```

```python
plt.bar(df['batsman'],df['2015'], width=0.5)
```

```python
plt.bar(np.arange(df.shape[0]) - 0.2, df['2015'], width=0.2, color='yellow')
plt.bar(np.arange(df.shape[0]), df['2016'], width=0.2, color='red')
plt.bar(np.arange(df.shape[0]) + 0.2, df['2017'], width=0.2, color='blue')

plt.xticks(np.arange(df.shape[0]), df['batsman'])
plt.show()
```

```python
np.arange(df.shape[0])
```

```python
# a problem
children = [10,20,40,10,30]
colors = ['red red red red red red','blue blue blue blue','green green green green green','yellow yellow yellow yellow ','pink pink pink pink']
plt.bar(colors, children, color='red')
```

```python
# solution
children = [10,20,40,10,30]
colors = ['red red red red red red','blue blue blue blue','green green green green green','yellow yellow yellow yellow ','pink pink pink pink']
plt.bar(colors, children, color='red')
plt.xticks(rotation='vertical')
```

```python
# Stacked Bar chart
plt.bar(df['batsman'], df['2017'], label='2017')
plt.bar(df['batsman'], df['2016'], bottom=(df['2017']), label='2016')
plt.bar(df['batsman'], df['2015'], bottom=(df['2016'] + df['2017']), label='2015')
plt.legend()
plt.show()
```

# Pie Chart
- Univariate/Bivariate Analysis
- Categorical vs numerical
- Use case - To find contibution on a standard scale

image.png

```python
# simple data
data = [23,45,100,20,49]
subjects = ['eng','science','maths','sst','hindi']
plt.pie(data, labels=subjects)

plt.show()
```

```python
# dataset
df = pd.read_csv('/content/gayle-175.csv')
df
```

```python
plt.pie(df['batsman_runs'], labels=df['batsman'], autopct='%0.1f%%')
plt.show()
```

```python
# percentage and colors
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',colors=['blue','green','yellow','pink','cyan','brown'])
plt.show()
```

```python
# explode shadow
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',explode=[0.3,0,0,0,0,0.1],shadow=True)
plt.show()
```
