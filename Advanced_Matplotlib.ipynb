```python
fig, ax = plt.subplots(nrows=2,ncols=1,sharex=True,figsize=(10,6))

ax[0].scatter(batters['avg'],batters['strike_rate'],color='red')
ax[1].scatter(batters['avg'],batters['runs'])

ax[0].set_title('Avg vs Strick rate')
ax[0].set_ylabel('Strick rate')

ax[1].set_title('Avg vs Runs')
ax[1].set_ylabel('runs')
ax[1].set_xlabel('Avg')
```

```python
# 4 by 4 subplot

fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(15,6))

ax[0,0].scatter(batters['avg'], batters['strike_rate'], color='red')
ax[0,1].scatter(batters['avg'], batters['runs'], color='red')

ax[1,0].hist(batters['avg'], color='blue')
ax[1,1].hist(batters['runs'], color='blue')

# ax[0,1].set_title('Avg vs Strick Rate')
# ax.set_xlabel('Avg vs Strick Rate')
```

```python
# 3rd way
fig = plt.figure()

ax1 = fig.add_subplot(2,1,1)   # 2 rows, 1 cols, first plot
ax1.scatter(batters['avg'],batters['strike_rate'],color='red')

ax2 = fig.add_subplot(2,1,2)
ax2.hist(batters['runs'])  
```

```python
fig = plt.figure()

ax1 = fig.add_subplot(1,2,1)   # 1 rows, 2 cols, first plot
ax1.scatter(batters['avg'],batters['strike_rate'],color='red')

ax2 = fig.add_subplot(1,2,2)
ax2.hist(batters['runs'])  
```

```python
fig = plt.figure()

ax1 = fig.add_subplot(2,2,1)   # 2 rows, 2 cols, first plot
ax1.scatter(batters['avg'],batters['strike_rate'],color='red')

ax2 = fig.add_subplot(2,2,2)   # 2 rows, 2 cols, second plot
ax2.hist(batters['runs'])  

ax3 = fig.add_subplot(2,2,3)   # 2 rows, 2 cols, third plot
ax3.hist(batters['avg'])  
```

# 3D Surface Plot

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)

xx, yy= np.meshgrid(x,y)
```

```python
z = xx**2 + yy**2
z.shape
```

```python
fig = plt.figure(figsize=(12,6))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z, cmap='viridis')
fig.colorbar(p)
```

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)

x, y= np.meshgrid(x,y)

z = np.sin(x) + np.cos(y)
fig = plt.figure(figsize=(14,6))
ax = plt.subplot(projection='3d')
p = ax.plot_surface(x,y,z, cmap='viridis')
fig.colorbar(p)
```

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)

x, y= np.meshgrid(x,y)
z = x**2 + y**2

fig = plt.figure(figsize=(12,6))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(x,y,z,cmap='viridis')
fig.colorbar(p)
```

# Contour Plots

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)

x, y= np.meshgrid(x,y)
z = x**2 + y**2

fig = plt.figure(figsize=(12,6))

ax = plt.subplot()

p = ax.contour(x,y,z,cmap='viridis')
fig.colorbar(p)
```

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)

x, y= np.meshgrid(x,y)
z = x**2 + y**2

fig = plt.figure(figsize=(12,6))

ax = plt.subplot()

p = ax.contourf(x,y,z,cmap='viridis')
fig.colorbar(p)
```

# Heatmap


```python
delivery = pd.read_csv('/content/IPL_Ball_by_Ball_2008_2022.csv')
delivery.head()
```

```python
temp_df = delivery[(delivery['ballnumber'].isin([1,2,3,4,5,6])) & (delivery['batsman_run']==6)]
```

```python
grid = temp_df.pivot_table(index='overs',columns='ballnumber',values='batsman_run',aggfunc='count')
```

```python
plt.figure(figsize=(10,6))
plt.imshow(grid)
```
