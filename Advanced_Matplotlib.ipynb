```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# colored Scaterplot

```python
iris = pd.read_csv('/content/iris.csv')
iris.head()
```

```python
iris.sample(5)
```

```python
iris['Species'] = iris['Species'].replace({'Iris-setosa':0, 'Iris-versicolor':1, 'Iris-virginica':2})
iris.sample(5)
```

```python
plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'], c=iris['Species'], cmap='jet')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
```

# plot Size


```python
plt.figure(figsize=(15,6))

plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'], c=iris['Species'], cmap='jet')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
```

```python
batters = pd.read_csv('/content/batter.csv')
batters.head()
```

```python
sample_df = batters.head(100).sample(25,random_state=5)
sample_df
```

```python
plt.figure(figsize=(18,10))



plt.scatter(sample_df['avg'], sample_df['strike_rate'], s=sample_df['runs'])
for i in range(sample_df.shape[0]):
  plt.text(sample_df['avg'].values[i], sample_df['strike_rate'].values[i], sample_df['batter'].values[i])
```

# Horizontal and Vertical lines


```python
plt.figure(figsize=(18,10))

plt.axhline(130, color='red')
plt.axvline(30, color='red')
plt.axhline(140, color='green')
plt;

plt.scatter(sample_df['avg'], sample_df['strike_rate'], s=sample_df['runs'])
for i in range(sample_df.shape[0]):
  plt.text(sample_df['avg'].values[i], sample_df['strike_rate'].values[i], sample_df['batter'].values[i])
```

# Subplots

```python
# A diff way to plot graphs
batters.head()
```

```python
plt.figure(figsize=(15,6))
plt.scatter(batters['avg'], batters['strike_rate'])
plt.title('Something')
plt.xlabel('Avg')
plt.ylabel('Strike Rate')

plt.show()
```

```python
# second way

fig, ax = plt.subplots(figsize=(15,6))
ax.scatter(batters['avg'], batters['strike_rate'], color='red', marker='+')
ax.set_title('Something')
ax.set_xlabel('avg')
ax.set_ylabel=('strike_rate')
fig.show()
```

```python
fig, ax = plt.subplots(nrows=2,ncols=1,sharex=True,figsize=(10,6))

ax[0].scatter(batters['avg'],batters['strike_rate'],color='red')
ax[1].scatter(batters['avg'],batters['runs'])

ax[0].set_title('Avg vs Strick rate')
ax[0].set_ylabel('Strick rate')

ax[1].set_title('Avg vs Runs')
ax[1].set_ylabel('runs')
ax[1].set_xlabel('Avg')
```

```python
# 4 by 4 subplot

fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(15,6))

ax[0,0].scatter(batters['avg'], batters['strike_rate'], color='red')
ax[0,1].scatter(batters['avg'], batters['runs'], color='red')

ax[1,0].hist(batters['avg'], color='blue')
ax[1,1].hist(batters['runs'], color='blue')

# ax[0,1].set_title('Avg vs Strick Rate')
# ax.set_xlabel('Avg vs Strick Rate')
```

```python
# 3rd way
fig = plt.figure()

ax1 = fig.add_subplot(2,1,1)   # 2 rows, 1 cols, first plot
ax1.scatter(batters['avg'],batters['strike_rate'],color='red')

ax2 = fig.add_subplot(2,1,2)
ax2.hist(batters['runs'])  
```

```python
fig = plt.figure()

ax1 = fig.add_subplot(1,2,1)   # 1 rows, 2 cols, first plot
ax1.scatter(batters['avg'],batters['strike_rate'],color='red')

ax2 = fig.add_subplot(1,2,2)
ax2.hist(batters['runs'])  
```

```python
fig = plt.figure()

ax1 = fig.add_subplot(2,2,1)   # 2 rows, 2 cols, first plot
ax1.scatter(batters['avg'],batters['strike_rate'],color='red')

ax2 = fig.add_subplot(2,2,2)   # 2 rows, 2 cols, second plot
ax2.hist(batters['runs'])  

ax3 = fig.add_subplot(2,2,3)   # 2 rows, 2 cols, third plot
ax3.hist(batters['avg'])  
```
